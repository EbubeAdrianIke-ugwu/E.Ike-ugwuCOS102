```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

def check_employee():
    name, dept = ent_name.get(), ent_dept.get()
    df = pd.read_csv('GIG-logistics.csv')

    # Find matching row
    match = df[(df['Name'] == name) & (df['Department'] == dept)]
    
    if not match.empty:
        # Get everyone else in that department
        others = df[(df['Department'] == dept) & (df['Name'] != name)]['Name'].tolist()
        messagebox.showinfo("Success", f"Welcome {name}!\nOthers: {', '.join(others)}")
    else:
        messagebox.showerror("Error", "Invalid Name or Department")

# --- UI Setup ---
root = tk.Tk()
root.title("GIG Check")

tk.Label(root, text="Name:").grid(row=0, column=0)
ent_name = tk.Entry(root)
ent_name.grid(row=0, column=1)

tk.Label(root, text="Dept:").grid(row=1, column=0)
ent_dept = tk.Entry(root)
ent_dept.grid(row=1, column=1)

tk.Button(root, text="Login", command=check_employee).grid(row=2, columnspan=2)

root.mainloop()
```
